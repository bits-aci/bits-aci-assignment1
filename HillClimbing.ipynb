```python
import numpy as np
import random

# Matrix representing the traversal grid
# 0 Represents Food supply center
# And other entries from 1 - 8 represent the distance of the tents from the Food supply center in clockwise
tsp=[
[0,2,4,7,8,10,13,15,16],
[2,0,2,5,6,8,11,13,14],
[4,2,0,3,4,6,9,11,12],
[7,5,3,0,1,3,6,8,9],
[8,6,4,1,0,2,5,7,8],
[10,8,6,3,2,0,3,5,6],
[13,11,9,6,5,3,0,2,3],
[15,13,11,8,7,5,2,0,1],
[16,14,12,9,8,6,3,1,0]
]
print(len(tsp))


# Picking random order of tents to visit
def randomSolution(tsp):
    cities = list(range(0,len(tsp)))
    print(cities);
    solution = []
    for i in range(0,len(tsp)-1):
      randomCity = cities[random.randint(1, len(cities)-1)]
      solution.append(randomCity)
      cities.remove(randomCity)
    print("solution before append: ",solution)
    solution.append(0)
    print("Solution first:", solution);
    return solution

solution = randomSolution(tsp);


# Define total distance travelled by the Agent for a given solution
def routeLength(tsp, solution):
    routeLength = 0
    print(range(0,len(solution)-1))
    for i in range(0,len(solution)-1):
        print("tsp location",solution[i - 1], solution[i])
        print("tsp values:", tsp[solution[i - 1]][solution[i]])
        ##
        if i == 0:
          cw = tsp[solution[i - 1]][solution[i]]
          print("cw: ",cw)
          acw = 18-tsp[solution[i - 1]][solution[i]]
          print("acw: ",acw)
          if cw < acw:
            routeLength += cw
          else:
            routeLength += acw
        ##
        print("routeLength:", routeLength)
        routeLength += tsp[solution[i - 1]][solution[i]]
    print("routeLength: ",routeLength)
    cw = tsp[0][solution[i]]
    print("cw: ",cw)
    acw = 18-tsp[0][solution[i]]
    print("acw: ",acw)
    if cw < acw:
     routeLength += cw
    else:
     routeLength += acw
    print("routeLength: ",routeLength)
    print("Solution:",solution)
    return routeLength

print(routeLength(tsp, solution))

# Function to get all possible neighbour tents
# def getNeighbours(solution):
#     neighbours = []
#     for i in range(len(solution)):
#         for j in range(i + 1, len(solution)):
#             neighbour = solution.copy()
#             neighbour[i] = solution[j]
#             neighbour[j] = solution[i]
#             neighbours.append(neighbour)
#             print("neighbours: ",neighbours)
#     return neighbours

def getNeighbours(solution):
    solution.remove(0)
    neighbours = []
    for i in range(len(solution)):
        for j in range(i + 1, len(solution)):
            neighbour = solution.copy()
            neighbour[i] = solution[j]
            neighbour[j] = solution[i]
            neighbour.append(0)
            neighbours.append(neighbour)
            print("neighbours: ",neighbours)
    return neighbours

# Function to get the best neighbour tents considering the total length of the route
def getBestNeighbour(tsp, neighbours):
    bestRouteLength = routeLength(tsp, neighbours[0])
    bestNeighbour = neighbours[0]
    for neighbour in neighbours:
        currentRouteLength = routeLength(tsp, neighbour)
        if currentRouteLength < bestRouteLength:
            bestRouteLength = currentRouteLength
            bestNeighbour = neighbour
    return bestNeighbour, bestRouteLength


# Iteration for the Hill climbing algorithm
def hillClimbing(tsp):
    currentSolution = randomSolution(tsp)
    currentRouteLength = routeLength(tsp, currentSolution)
    neighbours = getNeighbours(currentSolution)
    bestNeighbour, bestNeighbourRouteLength = getBestNeighbour(tsp, neighbours)

    while bestNeighbourRouteLength < currentRouteLength:
        currentSolution = bestNeighbour
        currentRouteLength = bestNeighbourRouteLength
        neighbours = getNeighbours(currentSolution)
        bestNeighbour, bestNeighbourRouteLength = getBestNeighbour(tsp, neighbours)
    return currentSolution, currentRouteLength

# Invoking Hill Climbing algorithm
# Printing out identified best solution and total length of the route
print(hillClimbing(tsp))
```
